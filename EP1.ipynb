```python
import numpy as np
import pandas as pd
```

```python
# Inicializamos as variaveis
NUSP = 9079703
A = [0, 1]
folhas = ['0', '01', '011', '111']
h = {'0':0.9, '01':0.7, '011':0.9, '111':0.7}

# Aqui, como não podemos ter indices negativos em arrays para alocar X(-3), X(-2) e X(-1),
# transladamos para um vetor de indices positivos.
# Assim, X tera 104 posicoes (contando com o zero) e a variavel t e m serão acrescidas de 3 tambem
X = [None]*1004
X[0] = 0
X[1] = 1
X[2] = 1
t = 3
m = 1003
```

```python
# Aqui, para facilitar a resolucao da questao 2, montei um dataset para exportar para o Excel
columns = ['t (t atual)', 'U (Unif(0,1))', 'folha (C)', 'alfabeto (X[t])']
df = pd.DataFrame(columns=columns)
df.head()
```

```python
len(df)
```

```python
# Algoritmo do enunciado da questao 1
while t <= m:
    i = 1
    
#     Inicializamos o primeiro elemento da folha
    possivel_folha = str(X[t-i])
    
#     Iremos ate o passado mais distante para achar a folha em que estamos
    while possivel_folha not in folhas:
        i += 1
        possivel_folha = str(X[t-i]) + possivel_folha
        
#     Aqui, nos encontramos a folha exata e a passamos para a variavel C
    C = possivel_folha
    
#     Geramos um numero aleatorio segundo a distribuicao Uniforme entre 0 e 1
    U = np.random.uniform(0, 1)
    
#     Aqui podemos identificar a probabilidade de ocorrer 0 ou 1 no alfabeto, achando a probabilidade da nossa variavel
#     aleatoria U ser menor que h[C]
    if U < h[C]:
        X[t] = 0
    else:
        X[t] = 1
    
#     Criei esse dataset para facilitar a continuidade do trabalho
    df.loc[len(df)] = [t, U, C, X[t]]
    t += 1

#     Printamos X
X
```
